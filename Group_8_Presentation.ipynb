# TASK 1: Connecting to Data

## Load Required Packages

```python
from dash import Dash, html
import dash_ag_grid as dag
import pandas as pd
```

```python
# Incorporate data
df = pd.read_csv(r'C:\Users\New\Downloads\Excel_Scrum_G8\Python_Visualization\ob_report_2026.csv')
df = df.drop(columns=['_id', 'Institution Name', 'Causative Agent-2', 'Date Declared Over'])
df['Date Outbreak Began'] = pd.to_datetime(df['Date Outbreak Began'])

# Initialize the app
app = Dash()

# App layout
app.layout = [
    html.Div(children='TASK 1: Connecting to Data'),
    dag.AgGrid(
        rowData=df.to_dict('records'),
        columnDefs=[{"field": i} for i in df.columns]
    )
]
```

```python
# Run the app
if __name__ == '__main__':
    app.run(debug=True)
```

# TASK 2 Visualize Histogram

```python
# Import packages
from dash import dcc
import plotly.express as px
```

```python
df_grouped = df.groupby(['Date Outbreak Began', 'Active']).size().reset_index(name='count')
# App layout
app.layout = [
    html.Div(children='TASK 2 Visualize Histogram'),

# Group by 'Date Outbreak Began' and 'Active' and count occurrences
# Count the number of outbreaks for each date and active status
    dcc.Graph(figure=px.bar(df_grouped, x='Date Outbreak Began', y='count', color='Active',
             title='Count of Outbreaks by Date and Active Status')
              # html.Hr()
              )
]
```

```python
# Run the app
if __name__ == '__main__':
    app.run(port = 8040)
```

# TASK 3 Controls and Callbacks

```python
# Import packages
from dash import callback, Output, Input
import dash
import more_itertools
```

```python
app = dash.Dash(__name__)

# App layout
app.layout = html.Div([
    html.Div(children='Outbreak Report Dashboard'),
    html.Hr(),
    dcc.Dropdown(
        id='chart-type',
        options=[
            {'label': 'Line', 'value': 'Line'},
            {'label': 'Treemap', 'value': 'Treemap'}
        ],
        value='Line',
    ),
      dcc.Graph(id='main-graph'),
    html.Div(id='output-info'),
])

# Callback to update the graph
@callback(
    Output('main-graph', 'figure'),
    [Input('chart-type', 'value'),
    #  Input('category-item', 'value')
     ]
)
def update_graph(chart_type):
    if chart_type == 'Line':
        dff = df.groupby('Date Outbreak Began').size().reset_index(name='Count')
        fig = px.line(dff, x='Date Outbreak Began', y='Count', title='Outbreaks Over Time')
        fig.update_traces(hovertemplate='Date: %{x}<br>Count: %{y}')
    elif chart_type == 'Treemap':
        dff = df.groupby(['Outbreak Setting', 'Type of Outbreak', 'Causative Agent-1']).size().reset_index(name='Count')
        fig = px.treemap(dff, path=['Outbreak Setting', 'Type of Outbreak', 'Causative Agent-1'], values='Count',
                         title='Treemap of Outbreaks')
        fig.update_traces(hovertemplate='%{label}<br>Count: %{value}')
    else:
        fig = px.scatter()  # Empty figure as fallback
    return fig

# Callback to display click data
@callback(
    Output('output-info', 'children'),
    Input('main-graph', 'clickData')
)
def display_click_data(click_data):
    if click_data:
        point = click_data['points'][0]
        label = point.get('label', point.get('x', 'Unknown'))
        value = point.get('value', point.get('y', 'Unknown'))
        customdata = point.get('customdata', '')
        color = point.get('marker.color', 'Unknown')
        active = '' if color == 'Unknown' else f', Active: {color}'  # Simplified, may need adjustment
        return f'You clicked on: {label}, Value: {value}{active}'
    else:
        return 'Click on a part of the chart to see details.'
```

```python
# Run the app
if __name__ == '__main__':
    app.run(port = 8030)
```

# Task 4 Styling Your App: HTML and CSS

```python
# Initialize the Dash app with external stylesheet
external_stylesheets = ['https://codepen.io/chriddyp/pen/bWLwgP.css']
app = dash.Dash(__name__, external_stylesheets=external_stylesheets)

# App layout using Divs with classNames from the stylesheet
app.layout = html.Div([
    html.Div([
        html.Div(
            html.H1('Outbreak Report Dashboard'),
            className='twelve columns',
            style={'textAlign': 'center'}
        )
    ], className='row', style={'marginBottom': '20px'}),

    html.Div([
        html.Hr(className='twelve columns')
    ], className='row', style={'marginBottom': '20px'}),

    html.Div([
        html.Div(
            dcc.Dropdown(
                id='chart-type',
                options=[
                    {'label': 'Line', 'value': 'Line'},
                    {'label': 'Bar', 'value': 'Bar'},
                    {'label': 'Pie', 'value': 'Pie'},
                    {'label': 'Treemap', 'value': 'Treemap'}
                ],
                value='Bar',
                style={'width': '100%'}
            ),
            className='six columns'
        )
    ], className='row', style={'marginBottom': '20px'}),

    html.Div([
        html.Div(
            dcc.RadioItems(
                options=['Outbreak Setting', 'Type of Outbreak', 'Causative Agent-1'],
                value='Outbreak Setting',
                id='category-item',
                inline=True
            ),
            className='twelve columns'
        )
    ], className='row', style={'marginBottom': '20px'}),

    html.Div([
        html.Div(
            dcc.Graph(id='main-graph'),
            className='twelve columns'
        )
    ], className='row', style={'marginBottom': '20px'}),

    html.Div([
        html.Div(
            html.Div(id='output-info', style={'fontSize': '16px'}),
            className='twelve columns'
        )
    ], className='row')
], style={'padding': '10px'})

# Callback to update the graph
@callback(
    Output('main-graph', 'figure'),
    [Input('chart-type', 'value'),
     Input('category-item', 'value')]
)
def update_graph(chart_type, category):
    if chart_type == 'Line':
        dff = df.groupby('Date Outbreak Began').size().reset_index(name='Count')
        fig = px.line(dff, x='Date Outbreak Began', y='Count', title='Outbreaks Over Time')
        fig.update_traces(hovertemplate='Date: %{x}<br>Count: %{y}')
    elif chart_type == 'Bar':
        dff = df.groupby([category, 'Active']).size().reset_index(name='Count')
        fig = px.bar(dff, x=category, y='Count', color='Active', barmode='group',
                     title=f'Counts by {category} Grouped by Active')
        fig.update_traces(hovertemplate='Category: %{x}<br>Count: %{y}<br>Active: %{fullData.name}')
    elif chart_type == 'Pie':
        dff = df[category].value_counts().reset_index(name='count')
        fig = px.pie(dff, values='count', names=category, title=f'Distribution of {category}')
        fig.update_traces(hovertemplate='%{label}<br>Count: %{value}')
    elif chart_type == 'Treemap':
        dff = df.groupby(['Outbreak Setting', 'Type of Outbreak', 'Causative Agent-1']).size().reset_index(name='Count')
        fig = px.treemap(dff, path=['Outbreak Setting', 'Type of Outbreak', 'Causative Agent-1'], values='Count',
                         title='Treemap of Outbreaks')
        fig.update_traces(hovertemplate='%{label}<br>Count: %{value}')
    return fig

# Callback to display click data
@callback(
    Output('output-info', 'children'),
    Input('main-graph', 'clickData')
)
def display_click_data(click_data):
    if click_data:
        point = click_data['points'][0]
        label = point.get('label', point.get('x', 'Unknown'))
        value = point.get('value', point.get('y', 'Unknown'))
        # For bar chart, use legendgroup for Active status
        active = point.get('legendgroup', 'Unknown')
        active_text = f', Active: {active}' if active != 'Unknown' else ''
        return f'You clicked on: {label}, Value: {value}{active_text}'
    else:
        return 'Click on a part of the chart to see details.'
```

```python
# Run the app
if __name__ == '__main__':
    app.run(port = 8070)
```
